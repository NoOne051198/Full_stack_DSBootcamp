<font size ="3">__5. Write a function which will take list as a input and give me concatenation of all the element as and output.__</font><br>

```python
def list_concat(*l):
    ''' This function returns the concatenation of all entered'''
    d = []
    for i in l:
        if type(i)==list:
            d = d + i
    return d
```

<font size ="3">__11. Write a function which can read video file and play for you.__</font><br>

```python
def Fun11(p):
    '''This function play any mp4 video,
    Give the path of the video
    '''
    import cv2 # import the OpenCV library

    cap = cv2.VideoCapture(p)

    while(cap.isOpened()):
        ret,frame = cap.read()
        if not ret: break # break if no next frame

        #frame = cv2.resize(frame,(700,940))
        cv2.imshow("video",frame)   # show frame 

        if cv2.waitKey(10) & 0xFF == ord('q'):  # On press of q break
            break

    cap.release()
    cv2.destroyAllWindows()
```

<font size ="3">__14. Write a function to read a complete PDF file.__</font><br>

```python
def read_pdf(p):
    '''This function is able to read PDF file ,
    p  path of the PDF file '''
    # importing requirement modeles
    from PyPDF2 import PdfReader

    # creating a pdf reader object
    reader = PdfReader(p)

    # printing number of pages in pdf file
    c = len(reader.pages)

    # getting a specific page from the pdf file
    for i in range(c):
        page = reader.pages[i]

        # extracting text from page
        text = page.extract_text()
        print(text)
```

```python
read_pdf("G:\\000_photos&Docs\\job preperation\\cv's\\Excel_included.pdf")
```

<font size ="3">__15. Write a function to read a word file.__</font><br>

```python
def read_word(p):
    '''This function can read the content of your word document,
    privide the path of the word file'''
    import docx
    doc = docx.Document(p)
    all_paras = doc.paragraphs
    #print("There are total ",len(all_paras)," paragraph in your file \n paragraph shown one by one")
    print('----------')
    for para in all_paras:
        print(para.text)
        print('----------')
```
